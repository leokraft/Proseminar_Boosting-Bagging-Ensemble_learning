```python
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
from sklearn import tree

from dataset import load_stroke_prediction_dataset
from metrics import getMetrics
```

```python
# load the dataset and create a train test split
X, y = load_stroke_prediction_dataset()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# normalize the data
sc = StandardScaler()
X_train[:, 18:21] = sc.fit_transform(X_train[:, 18:21])
X_test[:, 18:21] = sc.transform(X_test[:, 18:21])
```

```python
# create a large descision tree, a small descision tree and random forest ensemble
# calculate accuracies and print them to compare the different methods

random_forest = RandomForestClassifier(n_estimators=10, max_depth=10, random_state=0)
print(f"Random forest {random_forest}:")
getMetrics(random_forest, X_train, X_test, y_train, y_test)

small_tree = tree.DecisionTreeClassifier(max_depth=10, random_state=0)
print(f"Small tree {small_tree}:")
getMetrics(small_tree, X_train, X_test, y_train, y_test)

large_tree = tree.DecisionTreeClassifier(max_depth=100, random_state=0)
print(f"Large tree {large_tree}:")
getMetrics(large_tree, X_train, X_test, y_train, y_test)
```
