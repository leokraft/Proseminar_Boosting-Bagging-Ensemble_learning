```python
from sklearn.model_selection import train_test_split

from sklearn.ensemble import AdaBoostClassifier
from sklearn import tree

from dataset import load_pima_indians_diabetes_dataset
from metrics import getMetrics
```

```python
# load the dataset and create a train test split
X, y = load_pima_indians_diabetes_dataset()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# create a large descision tree, a descision stump and adaboost ensemble
# calculate accuracies and print them to compare the different methods

large_tree = tree.DecisionTreeClassifier(max_depth=100, random_state=0)
print(f"Large tree  {large_tree}:")
getMetrics(large_tree, X_train, X_test, y_train, y_test)

descision_stump = tree.DecisionTreeClassifier(max_depth=1,  random_state=0)
print(f"Descision stump {descision_stump}:")
getMetrics(descision_stump, X_train, X_test, y_train, y_test)

# Default: The base estimator is DecisionTreeClassifier initialized with max_depth=1
adaboost = AdaBoostClassifier(tree.DecisionTreeClassifier(max_depth=1), n_estimators=100, random_state=0)
print(f"AdaBoost {adaboost}:")
getMetrics(adaboost, X_train, X_test, y_train, y_test)
```
